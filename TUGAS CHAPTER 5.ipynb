```python
import pandas as pd #untuk membaca dan mengolah data tabel (Excel)
from scipy.stats.mstats import gmean, hmean #menghitung geometric mean dan harmonic mean
import matplotlib.pyplot as plt #untuk membuat grafik

my_dataset = pd.read_excel('Smith_glass_post_NYT_data.xlsx', sheet_name='Supp_traces') #membaca file excel

a_mean = my_dataset.Zr.mean() #Menghitung rata-rata biasa dari kolom Zr
g_mean = gmean(my_dataset['Zr']) #Menghitung rata-rata geometri dari kolom Zr
h_mean = hmean(my_dataset['Zr']) #Menghitung rata-rata harmonik dari kolom Zr

print ('-------') #menampilkan teks
print ('arithmetic mean')
print ("{0:.1f} [ppm]".format(a_mean)) #satuan (parts per million)
print ('-------')

print ('geometric mean')
print ("{0:.1f} [ppm]".format(g_mean))
print ('-------')

print ('harmonic mean')
print ("{0:.1f} [ppm]".format(h_mean))
print ('-------')

fig, ax = plt.subplots() #membuat figure (area grafik)
ax.hist(my_dataset.Zr, bins='auto', density=True, edgecolor='k', label='Measurements Hist', alpha=0.8) #Menampilkan distribusi data
ax.axvline(a_mean, color='pink', label='Arithmetic mean', linewidth=2) #Menggambar garis vertikal pada nilai mean
ax.axvline(g_mean, color='purple',  label='Geometric mean', linewidth=2) 
ax.axvline(h_mean, color='blue',  label='Harmonic mean', linewidth=2)
ax.set_xlabel('Zr [ppm]') #memberi nama sumbu
ax.set_ylabel('Probability density') #memberi nama sumbu
ax.legend() #menampilkan legenda

```

```python
import pandas as pd #membaca dan mengolah data
import matplotlib.pyplot as plt #membuat visualisasi grafik

my_dataset = pd.read_excel('Smith_glass_post_NYT_data.xlsx', sheet_name='Supp_traces') #membaca data

variance =  my_dataset['Hf'].var() #Menghitung varians dari data Hf
stddev =  my_dataset['Hf'].std() #Menghitung simpangan baku


print ('-------') #menampilkan hasil
print ('Variance') #menampilkan hasil variansi
print("{0:.0f} [square ppm]".format(variance))
print ('-------')
print ('Standard Deviation') #menampilkan hasil simpangan baku
print("{0:.0f} [ppm]".format(stddev))
print ('-------')

fig, ax = plt.subplots() #membuat kanvas dan area plot
ax.hist(my_dataset.Hf, bins= 20, density = True, edgecolor='k', label='Measurements Hist') #Menampilkan distribusi data Hf
ax.axvline(my_dataset['Hf'].mean()-stddev, color='purple', label=r'mean - 1$\sigma$', linewidth=2) #menunjukan garis mean
ax.axvline(my_dataset['Hf'].mean()+stddev, color='pink', label=r'mean + 1$\sigma$', linewidth=2) #menunjukan garis mean
ax.axvspan(my_dataset['Hf'].mean()-stddev, my_dataset['Hf'].mean()+stddev, alpha=0.1, color='orange', label=r'mean $\pm$ 1$\sigma$') #memberi area arsiran (warna transparan) pada grafik di antara dua nilai pada sumbu X.
ax.set_xlabel('Hf [ppm]') #memberi nama label sumbu x
ax.set_ylabel('Probability density') #memberi nama label sumbu y
ax.legend() #menampilkan legenda
```

```python
import pandas as pd #membaca data Excel
import numpy as np #enghitung statistik (persentil)
import matplotlib.pyplot as plt #membuat grafik

my_dataset = pd.read_excel('Smith_glass_post_NYT_data.xlsx', sheet_name='Supp_traces') #membaca file data  excel

q1 = np.percentile(my_dataset.Zr, 25, method = 'midpoint') #menghitung kuartil pertama
q3 = np.percentile(my_dataset.Zr, 75, method = 'midpoint') #menghitung kuartil ketiga
  
iqr = q3 - q1 # Interquaritle range (IQR)  #Menghitung jarak antara Q3 dan Q1

print ('-------') #menampilkan hasil
print ('Interquaritle range (IQR): {0:.0f} [ppm]'.format(iqr))
print ('-------')

fig, ax = plt.subplots() #membuat grafik area plot
ax.hist(my_dataset.Zr, bins='auto', density=True, edgecolor='k', label='Measurements Hist') #Menampilkan distribusi data Zr
ax.axvline(q1, color='purple', label='Q1', linewidth=2) #memberi warna pada kuartil 1
ax.axvline(q3, color='pink', label='Q3', linewidth=2) #memberi warna pada kuartil tiga
ax.axvspan(q1, q3, alpha=0.1, color='orange', label='Interquaritle range (IQR)') #Memberi arsiran antara Q1 dan Q3
ax.set_xlabel('Zr [ppm]') #menamai label sumbu x
ax.set_ylabel('Probability density') #menamai label sumbu y
ax.legend() #menampilkan legenda
```

```python
import pandas as pd #membaca data
import numpy as np #menghitung histogram dan modus
import matplotlib.pyplot as plt #membuat visualisasi grafik

my_dataset = pd.read_excel('Smith_glass_post_NYT_data.xlsx', sheet_name='Supp_traces') #membaca data excel

a_mean = my_dataset.U.mean() #Menghitung rata-rata biasa dari data U

median = my_dataset.U.median() #Mengambil nilai tengah dari data U

hist, bin_edges = np.histogram(my_dataset['Zr'], bins=20, density=True) #Membuat histogram numerik
modal_value = (bin_edges[hist.argmax()] +  bin_edges[hist.argmax()+1])/2 #mencari bin dengan frekuensi tertinggi

fig, ax = plt.subplots() #Membuat kanvas dan area plot
ax.hist(my_dataset.U, bins=20, density=True, edgecolor='k', label="Measurements Hist") #Menampilkan distribusi data U
ax.axvline(modal_value, color='pink', label='Modal Value', linewidth=2) #Menampilkan posisi modus (nilai paling sering muncul)
ax.axvline(median, color='purple',  label='Median Value', linewidth=2) #Menampilkan posisi median
ax.axvline(a_mean, color='maroon',  label='Arithmetic mean', linewidth=2) #Menampilkan posisi me
ax.set_xlabel('U [ppm]') #Memberi nama sumbu x
ax.set_ylabel('Probability density') #Memberi nama sumbu y
ax.legend() #menampilkan legenda
```

```python
import numpy as np #membantu hitung histogram (untuk modus)
from scipy.stats import skew #menghitung koefisien skewness (Fisher-Pearson)

a_mean = my_dataset.Er.mean() #menghitung rata-rata data Er
median = my_dataset.Er.median() #menghitung nilai tengah data Er
hist, bin_edges = np.histogram(my_dataset['Er'], bins=20, density=True) #mencari bin dengan frekuensi tinggi
modal_value = (bin_edges[hist.argmax()] + bin_edges[hist.argmax()+1])/2 #mengambil titik tengah
standard_deviation = my_dataset['Er'].std() #Mengukur sebaran data

a1 = (a_mean - modal_value) / standard_deviation #Mengukur skewness berdasarkan mean dan modus
a2 = 3 * (a_mean - median) / standard_deviation #Mengukur skewness berdasarkan mean dan median
g1 = skew(my_dataset['Er']) #Menghitung skewness menggunakan metode statistik standar

print ('-------') #menampilkan hasil
print ("Pearson's first coefficient of skewness: {:.2f}".format(a1))
print ("Pearson's 2nd moment of skewness: {:.2f}".format(a2))
print ("Fisher-Pearson's coefficient of skewness: {:.2f}".format(g1))
print ('-------')
```

```python
import pandas as pd #membaca dan mengolah data excel
import matplotlib.pyplot as plt #visualisasi grafik

my_dataset = pd.read_excel('Smith_glass_post_NYT_data.xlsx', sheet_name='Supp_traces') #membuka file excel dan membaca data

fig, ax = plt.subplots() #Membuat kanvas dan axes
my_flierprops = dict(markerfacecolor='#f8e9a1', markeredgecolor='#24305e', marker='o') #tampilan outlier (marker, warna)
my_medianprops = dict(color='#ff00ff', linewidth = 2) #tampilan garis median (warna, ketebalan)
my_boxprops = dict(facecolor='#a8d0e6', edgecolor='#24305e') #tampilan kotak boxplot (warna isi, warna tepi)
ax.boxplot(my_dataset.U, patch_artist = True, notch=True, flierprops=my_flierprops, medianprops=my_medianprops, boxprops=my_boxprops) #membuat boxplot
ax.set_ylabel('U [ppm]') #memberi nama sumbu y
ax.set_xticks([1]) #memberi nama sumbu y
ax.set_xticklabels(['all Epochs'])
plt.show() #Menampilkan boxplot di layar
```
