```python
import numpy as np #mengimport modul numpy yang digunakan untuk operasi numerik, khususnya array dan fungsi matematika

def integrate_trap(f, a, b, n): #membuat fungsi
    # Implementation of the trapezoidal rule
    h = (b-a)/n #Menghitung lebar tiap bagian (interval)
    x = np.linspace(a, b, n+1) #Membuat titik-titik dari a sampai b sebanyak n+1
    i=1 #Mulai perulangan dari indeks ke-1
    area = h*(f(x[0]) + f(x[n]))/2 #Menghitung bagian awal luas trapesium: titik pertama (x[0]), titik terakhir (x[n])
    while i<n: #Perulangan untuk menghitung titik-titik di tengah
        sup_rect = f(x[i])*h #Menghitung kontribusi luas dari titik tengah
        area += sup_rect  #Menambahkan ke total luas
        i += 1 #Menambah indeks agar lanjut ke titik berikutnya
    return area #Mengembalikan hasil total luas

'''
We test the trapezoidal rule on the known sine funcion were the  
definite integral in the interval [0, %*\textcolor{codeviolet}{$\pi$}*)/2] is equal to 1. 
'''

sup_5 = integrate_trap(lambda x: x**2, 0, 1, 5) #Menghitung integral fungsi x² dari 0 ke 1 dengan 5 bagian
sup_10 = integrate_trap(lambda x : x**2, 0, 1, 10) #Menghitung integral fungsi x² dari 0 ke 1 dengan 10 bagian

print('Using n=5, the trapezoidal rule returns a value of {:.2f}'.format(sup_5)) #menampilkan hasil ke layar
print('Using n=10, the trapezoidal rule returns a value of {:.2f}'.format(sup_10))
   
```

```python
def simple_lithopress(z,ro=3000, g=9.8): #Membuat fungsi bernama simple_lithopress, z = kedalaman (m), ro = densitas batuan (kg/m³), default = 3000, g = percepatan gravitasi (m/s²), default = 9.8
    pressure = z*g*ro/1e6 # return the pressure in MPa #Menghitung tekanan di kedalaman z
    return pressure #Mengembalikan hasil tekanan dari fungsi

my_pressure = simple_lithopress(z=2000) #menjalankan fungsi untuk kedalaman 2000 m
print('The pressure at 2000 meters is {0:.0f} MPa'.format(my_pressure)) #menampilkan hasil tekanan ke layar
```

```python
import numpy as np # Mengimpor library NumPy untuk pengolahan array
import matplotlib.pyplot as plt # Mengimpor library Matplotlib untuk visualisasi grafik
from scipy.integrate import trapezoid # Mengimpor fungsi trapz untuk integrasi numerik

def hitung_tekanan(r_km, ro_kg, g_ms2): # Mendefinisikan fungsi tekanan dengan input radius, densitas, dan gravitasi
    p = np.zeros(len(r_km)) # Membuat array kosong p berisi nol sepanjang array r_km
    r_m = r_km * 1000 # Mengonversi radius dari Kilometer ke Meter untuk standar satuan SI
    
    for i in range(0, len(r_m)): # Melakukan perulangan untuk setiap titik kedalaman i
        r1 = r_m[i:len(r_m)] # Mengambil potongan radius dari titik i hingga ke permukaan (titik terluar)
        ro1 = ro_kg[i:len(r_m)] # Mengambil potongan densitas dari titik i hingga ke permukaan
        g1 = g_ms2[i:len(g_ms2)] # Mengambil potongan gravitasi dari titik i hingga ke permukaan
        
        y = ro1 * g1 # Menghitung fungsi integran (densitas dikali gravitasi)
        p1 = trapezoid(y, r1) # Mengintegralkan hasil kali rho*g terhadap perubahan radius (r)
        p[i] = abs(p1) # Menyimpan hasil integral (tekanan) ke dalam array p (menggunakan abs agar positif)
        
    return p # Mengembalikan array tekanan yang telah dihitung


radius = np.linspace(0.1, 6371, 500) # Membuat data radius dari 0.1 km sampai 6371 km
densitas = 13000 - (10000 * (radius / 6371)**2) # Membuat variabel densitas tidak konstan agar grafik melengkung
gravitasi = np.linspace(0, 9.8, 500) # Membuat asumsi nilai gravitasi dari pusat ke permukaan


p_pascal = hitung_tekanan(radius, densitas, gravitasi) # Memanggil fungsi untuk mendapatkan tekanan dalam Pascal
p_gpa = p_pascal / 1e9 # Mengonversi hasil tekanan dari Pascal ke Gigapascal (GPa)
z_kedalaman = 6371 - radius # Mengubah variabel radius menjadi kedalaman (permukaan = 0 km)

fig, ax = plt.subplots() # Membuat frame gambar dan sumbu grafik
ax.plot(z_kedalaman, p_gpa, color='navy') # Menggambar grafik kedalaman (x) terhadap tekanan (y) warna hijau
ax.grid(True) # Menampilkan garis bantu (grid) pada grafik
ax.set_ylabel('P [GPa]') # Memberi label sumbu Y sebagai Tekanan dalam GPa
ax.set_xlabel('Depth z from the Earth Surface [km]') # Memberi label sumbu X sebagai Kedalaman dalam km
plt.show() # Menampilkan hasil akhir grafik ke layar
```
